# LLM Local

Usando seu próprio servidor de AI generativa

A Wolfram Research, criadora do Mathematica, Wolfram Language, Wolfram|Alpha, Wolfram System Modeler, etc...

Mantêm um benchmark de LLMs, com mais de 90 modelos:
https://www.wolfram.com/llm-benchmarking-project/



Quando a empresa não quer submemter seus dados para a nuvem, podemos utilizar modelos de LLM locais, com qualidade próxima do GPT-4, por exemplo...

Para prototipação, podemos textar modelos pequenos em nosso PC, ou servidores *on-premises*.

Veja os projetos:

Ollama:<br>
https://ollama.com/

Llamafile - Mozilla:<br>
https://future.mozilla.org/builders/news_insights/introducing-llamafile/

Com apenas 3 comandos no Linux é possível rodar seu ambiete LLM local seguro.

```bash
wget https://huggingface.co/Mozilla/llava-v1.5-7b-llamafile/resolve/main/llava-v1.5-7b-q4.llamafile?download=true
chmod +x llava-v1.5-7b-q4.llamafile
./llava-v1.5-7b-q4.llamafile
```

O desempenho depende da capacidade de processamento e memória do seu PC/servidor e GPU (Placa gráfica NVIDIA) é recomendado.

Instalei em uma VM Linux no MS Azure para teste...

Liberar portas no **firewall**: 8080 e 5000

Criei um server temoprário na nuvem para teste do **llamafile**
