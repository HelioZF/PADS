## Panel

```python
def gera_grafico(x, y, c, cmap):
    return br_pop_df.hvplot.scatter(x=x, y=y, c=c, cmap=cmap)
    
```

```python
import holoviews as hv
import panel as pn

pn.extension()
```

```python
cores = hv.plotting.list_cmaps()
kw = dict(x=variaveis, y=variaveis, c=variaveis, cmap=cores)

pn.interact(gera_grafico, **kw)
```

## Dashboards

```python
var_x = pn.widgets.Select(name='Eixo x', options=variaveis,)
var_y = pn.widgets.Select(name='Eixo y', options=variaveis,)
var_c = pn.widgets.Select(name='Cor:', options=variaveis,)

var_cmap = pn.widgets.Select(name='CMAP', options=cores)
```

```python
@pn.depends(var_x, var_y, var_c, var_cmap)
def gera_grafico(x, y, c, cmap):
    return br_pop_df.hvplot.scatter(x=x, y=y, c=c, cmap=cmap)
    
```

```python
@pn.depends(var_x, var_y, var_c)
def gera_head(x, y, c):
    return br_pop_df[[x,y,c]].head()
    
```

```python
view = pn.Row(
    pn.Column(var_x, var_y, var_c, var_cmap),
    pn.Column(pn.panel(gera_grafico, width=600),
    pn.panel(gera_head, width=600))
)
```

```python
import holoviews as hv
import hvplot.pandas
hv.extension('bokeh')

m_df = pd.read_csv("rais-2014-municipalities.csv")
 
m_df
 
fields = list(m_df.keys())
 
fields
 
combobox_x = pn.widgets.Select(name='Eixo x', options=fields,)
combobox_y = pn.widgets.Select(name='Eixo y', options=fields,)
combobox_c = pn.widgets.Select(name='Cor:', options=fields,)
combobox_cmap = pn.widgets.Select(name='CMAP', options=cores)
 
@pn.depends(combobox_x, combobox_y, combobox_c, combobox_cmap)
def gera_grafico(x, y, c, cmap):
    return m_df.hvplot.scatter(x=x, y=y, c=c, cmap=cmap)
 
@pn.depends(combobox_x, combobox_y, combobox_c)
def gera_head(x, y, c):
    return m_df[[x,y,c]].head()
 
gui_view = pn.Row(
    pn.Column(combobox_x, combobox_y, combobox_c, var_cmap),
    pn.Column(pn.panel(gera_grafico, width=600),
              pn.panel(gera_head, width=600))
)
 
gui_view
```
