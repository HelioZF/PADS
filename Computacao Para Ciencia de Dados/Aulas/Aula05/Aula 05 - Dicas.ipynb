### Data Warehouse e Data Lake

Qual a diferênça entre DW e DL:

Data Warehouse

- Tecnologia consolidada utilizada por empresas médias e granes
- Usa banco de dados relacional (SQL)
- Faz ETL (Extraction Transformation and Load) dos sistemas corporativos e dados externos
- Guarda tudo em tabelas organizadas (normalizadas - tidy) - Star Scheme
- É consumido via ferramentas corporativas
- Qualquer usuário da empresa consegue usar e gerar novas consultas e relatórios
- Pode ser acessado via planilha
- Normalmente interno no data center da empresa (rede local)
- BI, Data Mart, Data Mining
- Obtem dados dos sistemas transacionais da empresa (OLTP) e carrega durante a noite para o DW (OLAP)

Data Lake

- Tecnologia nova, mas não subistituio o BI, é outra coisa
- Arquivos RAW (arquivos nos formatos originais)
- É um grande file server na nuvem (Google, AWS, Azure, etc...) na AWS é o S3
- É acessado por cientistas de dados
- O ETL é feito pelo cientista de dados como part do process de data science
- É uma bagunça
- Tá na moda
- Big Data - Volumes GIGANTES de dados não estruturados e semi-estruturados
