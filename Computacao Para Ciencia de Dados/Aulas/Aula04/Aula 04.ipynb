# Aula 04 - Gráficos em Python (Matplotlib)

Matplotlib é uma das mais populares bibliotecas para construção de gráficos em Python. E também a base para outras bibliotecas mais recentes. 

Em suas origens, matplotlib buscava seguir um approach MATLAB para construção de gráficos. Posteriormente, buscou-se adotar um paradigma orientado a objetos para a sua construção.

Este é um dos motivos que você vai encontrar alguns códigos que seguem o estilo MATLAB (estruturado) e outros que seguem a orientação a objetos. É aconselhável que você conheça ambas as formas, mas que pratique a última. 

Para compreender o matplotlib é importante entender a anatomia de um gráfico em Python.

## O estilo MATLAB

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.linspace(0, 10, 100)

# criamos a figure
plt.figure()

# criamos o primeiro subplot 
plt.subplot(2, 1, 1) # (rows, columns, painel number)

# realizamos o plot no primeiro subplot
plt.plot(x, np.sin(x))

# criamos o segundo subplot
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

Essa abordagem é muito utilizada, principalmente pelos usuários que migraram de outras plataformas para o Python. 

Observe que ela é *stateful*, isto é, ela mantém o controle de qual é o subplot atual, sob o qual os comandos `plt.plot` são aplicados.

```python
# tamanho da figura
fig, ax = plt.subplots(figsize=(10, 8))
# título do gráfico
ax.set_title('Gráfico da Idade vs. Altura (em cm)',
             fontsize=20, fontstyle='italic')
# rótulos dos eixos X e Y
ax.set_xlabel('Idade (anos)', fontsize=16)
ax.set_ylabel('Altura (cm)', fontsize=16);
# ativar Grid
ax.grid(True)
# X e Y ticks
ax.set_xticks([i*5 for i in range(12)])
ax.tick_params(axis="x", labelsize=12)
ax.set_ylim(100,200)
ax.tick_params(axis="y", labelsize=12);
# vamos plotar os dados
ax.scatter(x = age, y = height, c = 'orange', s = 150, edgecolors = 'k')

# adicionar anotações na imagem
ax.text(x = 15, y = 105, s = 'Há crescimento até 20 anos,\n depois há uma certa estabilidade',
       fontsize=12, rotation=30, linespacing=2)
ax.text(x = 22, y = 185, s = 'Ninguém possui mais que 180 cm', fontsize=12);

# adicionar linhas
ax.vlines(x = 20, ymin = 100, ymax=180, linestyles='dashed', color='blue', lw=3)
ax.hlines(y = 180, xmin=0, xmax= 55, linestyles='dashed', color='red', lw=3);

# legendas
ax.legend(['Altura em cm'], loc='lower right', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 10, 100)
y1, y2, y3 = np.cos(x), np.sin(x+1), np.tan(x+3)
names = ['Signal 1 - cos', 'Signal 2 - sin', 'Signal 3 - tan']

fig, axes = plt.subplots(nrows=3, figsize=(10, 8))

for ax, y, name in zip(axes, (y1, y2, y3), names):
    ax.plot(x, y, color='blue')
    ax.set_title(name, fontsize=14)
    ax.set(xticks=[], yticks=[])
    ax.set_xlim((0,10))

plt.show()
```

## Random Walk (fill between)

Random Walk - [Passeio Aleatório](https://pt.wikipedia.org/wiki/Passeio_aleat%C3%B3rio) - [MathWorld](https://mathworld.wolfram.com/RandomWalk1-Dimensional.html)


```python
np.random.randn(20).cumsum()
```

```python
np.linspace(0, 10, 20)
```

```python
np.random.seed(1) #Pseudo Aleatórios

y = np.random.randn(100).cumsum()
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(10,8))

ax.plot(x, y)
plt.show()
```

```python
np.random.seed(1)

y = np.random.randn(100).cumsum()
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(10,8))

ax.fill_between(x, y, alpha=0.4, edgecolor='k')
plt.show()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
def model(x): return x * np.sin(x)


xdata = np.array([1, 3, 5, 6, 8, 10])
ydata = model(xdata)
```

```python
xdata
```

```python
gp = GaussianProcessRegressor() 
```

```python
xdata.reshape(-1, 1)
```

```python
gp.fit(xdata.reshape(-1, 1), ydata)
```

```python
xfit = np.linspace(0, 10, 1000)
yfit, std = gp.predict(xfit.reshape(-1, 1), return_std=True)
```

```python
xfit
```

```python
yfit
```

```python
dyfit = 2 * std # 95% confidence interval
```

```python
plt.figure(figsize=(15, 10))

plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color='gray',
   alpha=0.2,  label='95% confidence interval')

plt.legend(loc='upper left',  labelcolor='k')
plt.show()
```

## Matplotlib em 3D

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))

ax = Axes3D(fig)

X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)

def f(x, y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

ax.plot_surface(X, Y, f(X, Y), rstride=1, cstride=1)
```

## O quarteto de Anscombe

O Quarteto de Anscombe são quatro conjuntos de dados que têm estatísticas descritivas quase idênticas (como a média e a variância), mas que têm distribuições muito diferentes e aparências muito distintas quando exibidos graficamente. [Wikipedia](https://pt.wikipedia.org/wiki/Quarteto_de_Anscombe)

É um clássico em ciências dos dados...

O Seaborn tem diversos conuntos de dados (datasets) classicos disponíveis:
